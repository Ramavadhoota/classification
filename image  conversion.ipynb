```python
from PIL import Image
image_folder = r"C:\Users\hegde\OneDrive\Desktop\images" 
grayscale_folder = r"C:\Users\hegde\OneDrive\Desktop\image2"  
if not os.path.exists(grayscale_folder):
    os.makedirs(grayscale_folder)
categories = ['apple frits', 'bicycle', 'car', 'cat', 'elephant', 
              'mango frits', 'person', 'pizza', 'tiger', 'train', 
              'truck', 'banana'] 
for category in categories:
    category_path = os.path.join(image_folder, category) 
    if os.path.exists(category_path):
        grayscale_category_path = os.path.join(grayscale_folder, category)
        if not os.path.exists(grayscale_category_path):
            os.makedirs(grayscale_category_path)
        for filename in os.listdir(category_path):
            file_path = os.path.join(category_path, filename)            
            try:
                img = Image.open(file_path)
                grayscale_img = img.convert("L")
                grayscale_img.save(os.path.join(grayscale_category_path, filename))
                print(f"{filename} in '{category}' converted to grayscale and saved.")            
            except Exception as e:
                print(f"Skipping file '{filename}' in '{category}' as it could not be processed: {e}")
print("All images have been processed and saved in 'image2' folder.")
```

```python
import numpy as np
grayscale_folder = r"C:\Users\hegde\OneDrive\Desktop\image2" 
image_data = []
labels = []
for category in os.listdir(grayscale_folder):
    category_path = os.path.join(grayscale_folder, category)  
    if os.path.isdir(category_path): 
        for filename in os.listdir(category_path):
            file_path = os.path.join(category_path, filename)          
            try:
                img = Image.open(file_path)
                img = img.resize((64, 64))  
                img_array = np.array(img) / 255.0  
                image_data.append(img_array)
                labels.append(category)  
                print(f"Processed image: {filename} from category: {category}")            
            except Exception as e:
                print(f"Skipping file '{filename}' as it could not be processed: {e}")
image_data = np.array(image_data)
labels = np.array(labels)
print(f"Dataset created with {len(image_data)} images.")
```

```python
import pandas as pd
grayscale_folder = r"C:\Users\hegde\OneDrive\Desktop\image2"  
dataset_folder = r"C:\Users\hegde\OneDrive\Desktop\Dataset" 
image_data = []
labels = []
for category in os.listdir(grayscale_folder):
    category_path = os.path.join(grayscale_folder, category)  
    if os.path.isdir(category_path):  
        for filename in os.listdir(category_path):
            file_path = os.path.join(category_path, filename)            
            try:
                img = Image.open(file_path)
                img = img.resize((64, 64)) 
                img_array = np.array(img).flatten() / 255.0  
                image_data.append(img_array)
                labels.append(category)  
                print(f"Processed image: {filename} from category: {category}")          
            except Exception as e:
                print(f"Skipping file '{filename}' as it could not be processed: {e}")
image_data = np.array(image_data)
labels = np.array(labels)
dataset_csv = os.path.join(dataset_folder, "grayscale_images.csv")
df = pd.DataFrame(image_data)
df['label'] = labels
df.to_csv(dataset_csv, index=False)
print(f"Dataset saved as CSV to {dataset_csv}")
dataset_npz = os.path.join(dataset_folder, "grayscale_images.npz")
np.savez(dataset_npz, images=image_data, labels=labels)
print(f"Dataset saved as NPZ to {dataset_npz}")
```
